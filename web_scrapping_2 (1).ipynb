```python
#Importation de la bibliothèque BeautiflSoup
from bs4 import BeautifulSoup
```

```python
#Lecture de la page web 
with open('quotes.toscrape.html', 'r') as f: 
  source=f.read()
  
```

```python
#Instenciatation de la fonction BeautifulSoup
docum = BeautifulSoup(source, 'html.parser')
```

```python
#Affichons le titre de la page web
titre=docum.title
titre
```

```python
#AffIchons l'ensemble des occurences de la balise a
contenu=docum.find_all("a")
contenu
```

```python
#Créons une fonction qui retourne le nom des chansons
def chansons():
  nom_chansons=[chant.text.strip() for chant in contenu]
  return nom_chansons
```

```python
#Affichons l'ensemble des chansons qui se trouvent dans la page web
chansons()
```

```python
#Affichons l'ensemble des occurences des balises div
top_div=docum.find_all('div')
top_div
```

```python
#Créons une fonction qui retourne le nom des artistes
def nom_artistes():
  nom_artistes=[div_.text.strip() for div_ in top_div]
  return nom_artistes
```

```python
#Vérification des données qui se trouvent dans la balise "span"
info_span=docum.find_all('span')
info_span
```

```python
#Récupération des données qui se trouvent dans la balise "span"
info_span=[span_.text.strip() for span_ in top_span]
info_span
```

```python
#Vérification des occurences href qui se trouvent dans la balise a
info_href=docum.find_all("a",href=True)
info_href
```

```python
#Récupération de l'ensemble des liens qui se trouvent de la page web
recup_lien=[link['href'] for link in docum.find_all('a', href=True)]
recup_lien
```
